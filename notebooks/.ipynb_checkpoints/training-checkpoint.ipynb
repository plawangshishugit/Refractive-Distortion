```python
class UIEBDataset(Dataset):
    def __init__(self, raw_dir, reference_dir, transform=None):
        self.raw_images = sorted(os.listdir(raw_dir))
        self.reference_images = sorted(os.listdir(reference_dir))
        self.raw_dir = raw_dir
        self.reference_dir = reference_dir
        self.transform = transform

        assert len(self.raw_images) == len(self.reference_images), "Mismatch in dataset sizes!"

    def __len__(self):
        return len(self.raw_images)

    def __getitem__(self, idx):
        raw_path = os.path.join(self.raw_dir, self.raw_images[idx])
        reference_path = os.path.join(self.reference_dir, self.reference_images[idx])

        raw_img = Image.open(raw_path).convert("RGB")
        reference_img = Image.open(reference_path).convert("RGB")

        if self.transform:
            raw_img = self.transform(raw_img)
            reference_img = self.transform(reference_img)

        return raw_img, reference_img

# Define image transformations
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor()
])

# Load dataset
dataset = UIEBDataset("C:\Users\plawa\miniProject/data/UIEB/raw-890", "C:\Users\plawa\miniProject/data/UIEB/reference-890", transform=transform)
train_loader = DataLoader(dataset, batch_size=8, shuffle=True)

# Verify dataset
sample_raw, sample_ref = next(iter(train_loader))
print("Raw Image Shape:", sample_raw.shape)
print("Reference Image Shape:", sample_ref.shape)
```
